#  Eleições 2022 e desenvolvimento regional
Iremos explorar possíveis correlações entre os resultados das eleições presidenciais do 1o turno de 2022 e índices de desenvolvimento.

```python
import pandas as pd
import numpy as np
import plotly.express as px
```

## Carregando os resultados eleitorais por município
- Crie o DataFrame `votos` a partir do csv obtido do TSE, do mesmo jeito que realizado no Desafio Pandas.

```python
votos = pd.read_csv('votacao_candidato_munzona_2022_BR.csv', sep = ';', encoding='Latin-1')
display(votos)
```

## Calculando os votos percentuais
- Calcularemos, a partir da tabela `votos` os resultados percentuais de cada candidato em cada município.
- Para esto, somaremos os votos de cada candidato em cada município (pois alguns municípios estão divididos em 'zonas') e logo dividiremos os valores pelo número total de votos de cada município.
- **SPOILER:** Como ajuda, o código a seguir dá uma solução possível para essa tarefa. Podem não olhar para ela e tentar descobrir como obter aquelas percentagens!

```python
# Somando os votos de cada candidato em cada município (pois tem municípios divididos em zonas)
# Aproveito para ficar só com algumas das colunas
# Com o método 'aggregate', escolho qual aggregate fazer para cada coluna após o groupby. 'first' fica com o primeiro valor (uso isso quando são todos iguais)
votos_mn = votos.groupby(['CD_MUNICIPIO','NR_CANDIDATO']).aggregate(
    {'SG_UF':'first', 
     'NM_MUNICIPIO':'first', 
     'NM_URNA_CANDIDATO':'first', 
     'QT_VOTOS_NOMINAIS_VALIDOS':'sum'})
display( votos_mn )
```

```python
# CD_MUNICIPIO e NR_CANDIDATO ficaram como index em votos_mn. Isso vai facilitar meu cálculo de %
# ... isto aqui funciona porque na hora de dividir ele sozinho vai alinhar os valores pelo index 'CD_MUNICIPIO'
votos_mn['PCT_VOTOS'] = 100 * votos_mn['QT_VOTOS_NOMINAIS_VALIDOS'] / votos_mn.groupby(['CD_MUNICIPIO'])['QT_VOTOS_NOMINAIS_VALIDOS'].sum()

display( votos_mn )
```

```python
# Agora vou resetar o index para deixar CD_MUNICIPIO e NR_CANDIDATO como colunas (quem preferir pode pular e deixar como index)
# ... aproveito para mudar a ordem das colunas para deixar elegante
votos_mn = votos_mn.reset_index()[['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO', 'NM_URNA_CANDIDATO', 'QT_VOTOS_NOMINAIS_VALIDOS', 'PCT_VOTOS']]

display(votos_mn)
```

# Cruzando a informação do TSE com a do IBGE
- Para contrastar os resultados eleitorais nos municípios com seus índices de desenvolvimento humano (IDH), cruzaremos nossa tabela `votos_mn` com uma de IDH por município (IDHM).

## Primeiro, adicionaremos os códigos IBGE dos municípios na tabela `votos_mn`
- TSE e IBGE não usam os mésmos códigos para identificar municípios, portanto usaremos primeiro uma tabela que permite convertir um no outro, obtida de: https://basedosdados.org/dataset/br-bd-diretorios-brasil?bdm_table=municipio.
    - Carregue a tabela `codigos_municipios.csv` em um DataFrame chamado `codigos_municipios`.
    - Observe a organização da informação nela (naquele link tem uma descrição de todas as colunas)

```python
codigos_municipios = pd.read_csv('codigos_municipios.csv', sep=',')
display(codigos_municipios)
```

- As colunas com os códigos TSE dos municípios tem nomes diferentes na tabela de votos e na de códigos. Usando os parámetros `left_on` e `right_on` do método `pd.merge()`, fusione `votos_mn` com `codigos_municipios` pelos códigos de município do TSE. Usando o parametro `how`, fique só com as líneas com códigos que existem nas duas tabelas.
    - Pode jogar fora colunas de `codigos_municipios` que não ache úteis antes do merge para não ficar com tantas colunas.
- Confira no olho que a fusão deu certo (por exemplo, que os nomes dos municípios que vem das duas tabelas batam).


```python
# Deixando só a tabela com os códigos e os nomes dos municípios (para posteriormente verificar se o resultado bateu)
codigos_municipios = codigos_municipios.iloc[:, [0,2,5]]
display(codigos_municipios)

# Fundindo as tabelas, reordenando e verificando o resultado.
# Note que usei o parâmetro inner e excluí as cidades que não 
# existem na base de dados do IDHM (principalmente as cidades de fora do Brasil)
votos_mn = pd.merge(votos_mn, codigos_municipios, left_on='CD_MUNICIPIO', right_on='id_municipio_tse', how='inner')
display(votos_mn)

# Removendo alguns índices que não serão utilizados daqui pra frente
# (Note também que os nomes dos municípios voltam a aparecer na tabela do IBGE)
votos_mn = votos_mn.drop(['CD_MUNICIPIO', 'id_municipio_tse', 'nome', 'NM_MUNICIPIO'], axis = 1)
display(votos_mn)
```

## Agora sim, cruzaremos a informação de votos por município com a de IDHM e população total.
O índice IDH é um número entre 0 e 1, quanto mais perto de 1 maior o 'desenvolvimento humano' no município. Existem 3 índices:
- IDHM_E: índice de educação.
- IDHM_L: índice de longevidade.
- IDHM_R: índice de renda.
- IDHM é a média desses 3.

Para mais informação: https://www.br.undp.org/content/brazil/pt/home/idh0/conceitos/o-que-e-o-idhm.html

- Carregue a tabela `IDHM_2010.csv` usando encoding 'Latin-1' em um DataFrame chamado `idhm`.
- Crie uma tabela `votos_idhm`, fusionando a tabela `votos_mn` com a `idhm` pelos códigos de município IBGE.
    - Note que, mais uma vez, as colunas com os códigos tem nome diferente nas duas tabelas.
- Olhe para o resultado. Agora em cada línea de `votos_idhm` temos informação sobre o voto em um município junto com informação de IDH e população total (coluna 'population') desse município.

```python
idhm = pd.read_csv('IDHM_2010.csv', sep=',', encoding='Latin-1')

# Fundindo as tabelas de idhm e de votos
votos_idhm = pd.merge(votos_mn, idhm, left_on='id_municipio', right_on='city_ibge_code', how='left')
display(votos_idhm)
```

### 1. Criando a tabela com os candidatos.

Para efetuar nossa análise, selecionaremos os dois candidatos com maior percentual de voto no 1° turno das eleições presidenciais de 2022: os candidatos Lula e Jair Bolsonaro.

```python
# Selecionando dois candidatos utilizando boolean masks
votos_idhm_correlacao = votos_idhm.loc[(votos_idhm['NR_CANDIDATO'] == 13) | (votos_idhm['NR_CANDIDATO'] == 22)]
display(votos_idhm_correlacao)
```

### 2. Criando o plot scatter dos dados obtidos

Aqui, plotamos um gráfico de scatter onde cada ponto representa um município. No eixo horizontal encontramos os valores do IDHM de cada município, enquanto no eixo vertical encontramos os valores percentuais de voto em cada município.

```python
fig = px.scatter(votos_idhm_correlacao, x='IDHM', y='PCT_VOTOS', color='NM_URNA_CANDIDATO',
                hover_data=['city', 'population', 'NR_CANDIDATO', 'SG_UF'],
                color_discrete_map={
                    'LULA': 'red',
                    'JAIR BOLSONARO': 'blue'
                }
                )
fig.update_traces(marker_size=4)
fig.show()
```

E podemos ver uma correlação interessante. Nota-se uma tendência linear com relação ao percentual de votos e o IDHM para cada candidato. Para o candidato Jair Bolsonaro a quantidade de votos parece aumentar de acordo com o índice do IDHM de cada município. Já para o candidato Lula temos o contrário.

Esta correlação pode indicar que municípios menos desenvolvidos (isto é, com índices menores de renda e longevidade, por exemplo) tendem a mostrar um percentual de votos maior para o candidato Lula em relação ao candidato Jair Bolsonaro.

### 3. Criando faixas de IDHM e plotando em barras os novos gráficos

Primeiramente, criaremos 10 faixas de IDHM para efetuarmos nossos cálculos. Note que utilizei a tabela que inicialmente continha todos os candidatos para determinar as faixas.

```python
faixa_idhm = pd.cut(votos_idhm['IDHM'], 10)
display(faixa_idhm)
```

Agora, criaremos uma tabela que correlaciona a porcentagens de votos de cada candidato de acordo com a faixa de IDHM. Para fazermos isso, calculamos a quantidade votais de votos por faixa de IDHM e a usamos para calcular a porcentagem de votos na tabela.

Em seguida, resetamos o index, renomeamos a coluna que contém a porcentagem de votos e transformamos o tipo do dado dos intervalos em string para que seja possível realizar o plot em barras.

```python
votos_totais_por_idhm = votos_idhm.groupby(faixa_idhm)['QT_VOTOS_NOMINAIS_VALIDOS'].sum()
display(votos_totais_por_idhm)

votos_idhm_faixas = votos_idhm_correlacao.groupby([faixa_idhm, 'NM_URNA_CANDIDATO'])['QT_VOTOS_NOMINAIS_VALIDOS'].sum()/votos_totais_por_idhm*100
votos_idhm_faixas = votos_idhm_faixas.reset_index()
votos_idhm_faixas = votos_idhm_faixas.rename(columns={'QT_VOTOS_NOMINAIS_VALIDOS': 'PCT_VOTOS'})
votos_idhm_faixas['IDHM'] = votos_idhm_faixas['IDHM'].astype('str')
display(votos_idhm_faixas)
```

Ainda, adicionaremos a informação que contém os votos do restante dos candidatos. Este valor pode ser facilmente calculado levando-se em conta que a soma das porcentagens deve ser sempre 100%.

```python
outros_tabela = votos_idhm_faixas.groupby('IDHM')['PCT_VOTOS'].sum()*(-1)+100
outros_tabela = outros_tabela.reset_index()
outros_tabela['NM_URNA_CANDIDATO'] = 'OUTROS'

votos_idhm_faixas = pd.merge(votos_idhm_faixas, outros_tabela, on=['IDHM', 'PCT_VOTOS', 'NM_URNA_CANDIDATO'], how='outer')
votos_idhm_faixas = votos_idhm_faixas.sort_values(by=['IDHM', 'NM_URNA_CANDIDATO'])
votos_idhm_faixas = votos_idhm_faixas.reset_index(drop=True)
display(votos_idhm_faixas)
```

Agora, plotaremos o resultado do gráfico em barras.

```python
fig = px.bar(votos_idhm_faixas, x='IDHM', y='PCT_VOTOS', color='NM_URNA_CANDIDATO') 
fig.show()
```

### 4. Criando faixas de população e plotando em barras os novos gráficos

Analogamente a como fizemos anteriormente, criaremos 7 intervalos de populações para efetuarmos nossos cálculos.

```python
intervalos_populacao = pd.cut(votos_idhm['population'], [100, 2500, 5000, 1e4, 1e5, 1e6, 1e7])
display(intervalos_populacao)
```

Agora, criaremos uma tabela que correlaciona a porcentagens de votos de cada candidato de acordo com a faixa populacional. Faremos isso de maneira também semelhante a como fizemos acima com uma única mudança: incluiremos o número de municípios contidos em cada faixa populacional para utilizarmos este dado no plot interativo.

```python
votos_totais_por_pop = votos_idhm.groupby(intervalos_populacao)['QT_VOTOS_NOMINAIS_VALIDOS'].sum()
display(votos_totais_por_pop)

numero_de_municipios_por_faixa = votos_idhm.groupby(intervalos_populacao)['id_municipio'].nunique()
numero_de_municipios_por_faixa = numero_de_municipios_por_faixa.rename('NUMERO_DE_MN')
display(numero_de_municipios_por_faixa)

votos_populacao = votos_idhm_correlacao.groupby([intervalos_populacao, 'NM_URNA_CANDIDATO'])['QT_VOTOS_NOMINAIS_VALIDOS'].sum()/votos_totais_por_pop*100
votos_populacao = votos_populacao.reset_index()
votos_populacao = pd.merge(votos_populacao, numero_de_municipios_por_faixa, on='population')
votos_populacao = votos_populacao.rename(columns={'QT_VOTOS_NOMINAIS_VALIDOS': 'PCT_VOTOS'})
display(votos_populacao)
```

```python
outros_tabela_pop = votos_populacao.groupby('population')['PCT_VOTOS'].sum()*(-1)+100
outros_tabela_pop = outros_tabela_pop.reset_index()
outros_tabela_pop['NM_URNA_CANDIDATO'] = 'OUTROS'
outros_tabela_pop = pd.merge(outros_tabela_pop, numero_de_municipios_por_faixa, on='population')
display(outros_tabela_pop)

votos_populacao = pd.merge(votos_populacao, outros_tabela_pop, on=['population', 'PCT_VOTOS', 'NM_URNA_CANDIDATO', 'NUMERO_DE_MN'], how='outer')
votos_populacao = votos_populacao.sort_values(by=['population', 'NM_URNA_CANDIDATO'])
votos_populacao = votos_populacao.reset_index(drop=True)

votos_populacao['population'] = votos_populacao['population'].astype('str')
display(votos_populacao)
```

Plotando o gráfico, finalmente:

```python
fig = px.bar(votos_populacao, x='population', y='PCT_VOTOS', color='NM_URNA_CANDIDATO', 
            hover_data=['NUMERO_DE_MN']
            ) 
fig.show()
```
