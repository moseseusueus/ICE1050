## Reduzindo a tabela
Tem muitas colunas que a gente não precisa, vamos deixar a tabela mais enxuta:
- Na tabela `votos`, fique só com as colunas: 'SG_UF','CD_MUNICIPIO','NM_MUNICIPIO','NR_ZONA','NR_CANDIDATO','NM_URNA_CANDIDATO' e 'QT_VOTOS_NOMINAIS_VALIDOS'

# Resultados eleitorais
Vamos usar `.groupby()` e o aggregate `.sum()` para somar os votos de cada candidato en diferentes regiões.
## Qual foi o resultado total no Brasil?
- Obtenha a soma total de votos de cada candidato no Brasil todo. Apresente o resultado em ordem descendente com `.sort_values()`
- Converta os valores em percentagens do total.
- Apresente o resultado como "pie plot" (`.plot.pie()`)

```python
from matplotlib.pyplot import legend

votos_totais = votos2['QT_VOTOS_NOMINAIS_VALIDOS'].sum()
display(votos_totais)

votos_percentuais = votos2.groupby(['NM_URNA_CANDIDATO','NR_CANDIDATO'])['QT_VOTOS_NOMINAIS_VALIDOS'].sum()/votos_totais*100
display(votos_percentuais)

votos_percentuais.plot.pie(figsize=(14,14), fontsize = 10, legend=True, autopct = '%.2f%%')
```

## Qual foi o resultado no RN?
- Repita o anterior, mas selecionando previamente só os votos do RN.

```python
from matplotlib.pyplot import legend

votos_totais_por_estado = votos2.groupby('SG_UF')['QT_VOTOS_NOMINAIS_VALIDOS'].sum()
votos_totais_RN = votos_totais_por_estado['RN']
display(votos_totais_RN)

votos_candidatos_por_estado = votos2.groupby(['NM_URNA_CANDIDATO','NR_CANDIDATO','SG_UF'])['QT_VOTOS_NOMINAIS_VALIDOS'].sum()
display(votos_candidatos_por_estado)

votos_percentuais_RN = votos_candidatos_por_estado[:,:,'RN']/votos_totais_RN*100
display(votos_percentuais_RN)

votos_percentuais_RN.plot.pie(figsize=(14,14), fontsize = 10, legend=True, autopct = '%.2f%%')
```
